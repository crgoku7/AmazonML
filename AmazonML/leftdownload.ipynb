```python
import os
import pandas as pd

# Read the CSV file
test_df = pd.read_csv('test.csv')

# Define the folders where images are stored
folders = ['folder_2', 'folder_3', 'folder_4', 'folder_5', 'folder_6', 'folder_7', 'folder_8', 'folder_9', 'folder_10']  # Update with your actual folder names

# List to store missing image paths
missing_images = []

# Set custom start/end index (relative to test.csv, starting from 13121 to 131188)
start_idx = 13121  # This is the serial number you want to start from
end_idx = 131188  # This is the last image index

# Loop through each row in the CSV within the specified range
for idx in range(start_idx, end_idx + 1):
    if idx > len(test_df):
        print(f"Skipping index {idx} as it exceeds DataFrame bounds")
        continue

    row = test_df.iloc[idx - 1]  # Adjust for 1-based indexing in the CSV (start from row 13121)

    # The image is saved with the index number starting from 2, so we use idx as the image name
    image_name = f"{idx}.jpg"  # Image filenames start from 2, so use the current idx directly
    image_found = False

    # Check if the image exists in any folder
    for folder in folders:
        full_image_path = os.path.join(folder, image_name)
        if os.path.exists(full_image_path):
            image_found = True
            break
    
    # If image is not found, store its path
    if not image_found:
        missing_images.append({
            'index': row['index'],          # From CSV
            'image path': row['image_link'],  # Keep the original image path
            'group id': row['group_id'],
            'entity_name': row['entity_name']
        })

# Convert the missing images list to a DataFrame and save to a new CSV
if missing_images:
    missing_df = pd.DataFrame(missing_images)
    missing_df.to_csv('missing_images.csv', index=False)
    print(f"Missing images saved to missing_images.csv")
else:
    print("No missing images found")
```

```python
import csv
import requests
import os
import urllib.parse
import math

def download_image(image_url, image_name):
    try:
        response = requests.get(image_url)
        if response.status_code == 200:
            folder_path = 'E:/AMAZONML/student_resource 3/src/missing_downloads'
            image_path = os.path.join(folder_path, image_name + '.jpg')
            with open(image_path, 'wb') as f:
                f.write(response.content)
            print(f"Downloaded: {image_name}")
        else:
            print(f"Failed to download: {image_url}")
    except Exception as e:
        print(f"Error downloading {image_url}: {e}")

def create_valid_filename(name):
    return urllib.parse.quote_plus(name)

csv_file = 'E:/AMAZONML/student_resource 3/missing_images.csv'  # Replace with your CSV file name

# Count total number of images in the CSV file
with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    total_images = sum(1 for row in csv_reader) - 1  # Subtract 1 for the header

# Calculate how many images per folder
images_per_folder = math.ceil((131287+1) / 10)

with open(csv_file, 'r') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)  # Skip the header

    for idx, row in enumerate(csv_reader):
        image_name = row[0]  # First column has the image name
        image_url = row[1]  # Second column has the image URL
        
        # Generate valid image name
        valid_image_name = create_valid_filename(image_name)
        
        # Download the image into the appropriate folder
        download_image(image_url, valid_image_name)
```
